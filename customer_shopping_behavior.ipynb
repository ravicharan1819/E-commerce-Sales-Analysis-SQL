```python
#how to fill the null values??
#so mean can get affected by outliers results in the skew the distribtion of the data
#median is robust to outliers# so we imptute the median if the same catagory
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform (lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
labels = ['young_adult', 'adult', 'middle-aged', 'aged']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(5)
```

```python
# create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(5)
```

```python
# Dropping promo code used column

df = df.drop('promo_code_used', axis=1)
```
